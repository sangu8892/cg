```python
import sys
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *

angle=0.0
translate_x=0.0
translate_y=0.0
scale_factor=1.0
def init():
    glClearColor(1.0,1.0,1.0,1.0)
    gluOrtho2D(-1.0,1.0,-1.0,1.0)
    glMatrixMode(GL_PROJECTION)
    glLoadIdentity()

def draw_square():
    glBegin(GL_POLYGON)
    glVertex2f(-.10,-.10)
    glVertex2f(.10,-.10)
    glVertex2f(.10,.10)
    glVertex2f(-.10,.10)
    glEnd()

def display():
    global angle,translate_x,translate_y,scale_factor
    glClear(GL_COLOR_BUFFER_BIT)
    glMatrixMode(GL_MODELVIEW)
    glLoadIdentity()

    glTranslatef(translate_x,translate_y,0.0)
    glRotatef(angle,0.0,0.0,1.0)
    glScalef(scale_factor,scale_factor,1.0)

    glColor3f(0.0,0.0,1.0)
    draw_square()
    glutSwapBuffers()

def keyboard(key,x,y):
    global angle,translate_x,translate_y,scale_factor
    if key==b'q':
        sys.exit()
    elif key==b'r':
        angle +=10
    elif key==b'a':
        translate_x -=0.1
    elif key==b'd':
        translate_x +=0.1
    elif key==b'w':
        translate_y +=0.1
    elif key==b's':
        translate_y -=0.1
    elif key==b'+':
        scale_factor +=1.25
    
    glutPostRedisplay()

def main():
    glutInit(sys.argv)
    glutInitWindowSize(500,500)
    glutCreateWindow(b"2d transformation")
    glutDisplayFunc(display)
    glutKeyboardFunc(keyboard)
    init()
    glutMainLoop()

if __name__=="__main__":
    main()
    
    
    
```
